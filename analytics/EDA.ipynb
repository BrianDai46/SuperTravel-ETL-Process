# EDA

```python
travel['car_id'] = travel['car_id'].fillna(-1).astype(int)
travel['flight_id'] = travel['flight_id'].fillna(-1).astype(int)
travel['hotel_id'] = travel['hotel_id'].fillna(-1).astype(int)
travel['airbnb_id'] = travel['airbnb_id'].fillna(-1).astype(int)
travel['Transportation type'] = travel['Transportation type'].replace(['Flight', 'plane'], 'Plane')
travel['Transportation type'] = travel['Transportation type'].replace(['Car ', ' Car'], 'Car')
travel
```

# Average Costs by City

```python
#Average Cost by City
travel['per_capita_cost'] = travel['Accommodation cost'] + travel['Transportation cost']
grouped_data = travel.groupby('Destination_City').mean()[['Accommodation cost', 'Transportation cost', 'per_capita_cost']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.figure(figsize=(50, 5))
# Accommodation cost by city
sorted_grouped_data = grouped_data.sort_values('Accommodation cost', ascending=False)
plt.subplot(1, 5, 2)
sns.barplot(x=sorted_grouped_data.index, y=sorted_grouped_data['Accommodation cost'])
plt.title('Average Accommodation Cost by City')
plt.xticks(rotation=90)
plt.ylabel('Average Cost')
```

```python
plt.figure(figsize=(50, 5))
# Transportation cost by city
sorted_grouped_data1 = grouped_data.sort_values('Transportation cost', ascending=False)
plt.subplot(1, 5, 2)
sns.barplot(x=sorted_grouped_data1.index, y=sorted_grouped_data1['Transportation cost'])
plt.title('Average Transportation Cost by City')
plt.xticks(rotation=90)
plt.ylabel('Average Cost')
```

```python
plt.figure(figsize=(50, 5))
# Per capita cost by city
sorted_grouped_data2 = grouped_data.sort_values('per_capita_cost', ascending=False)
plt.subplot(1, 5, 2)
sns.barplot(x=sorted_grouped_data2.index, y=sorted_grouped_data2['per_capita_cost'])
plt.title('Average Per Capita Cost by City')
plt.xticks(rotation=90)
plt.ylabel('Average Cost')
```

# Average Costs by City and Type

```python
grouped_transport = travel.groupby(['Destination_City', 'Transportation type']).mean()['Transportation cost'].reset_index()
plt.figure(figsize=(50, 20))

# Transportation cost by city and type
ax = sns.barplot(x='Destination_City', y='Transportation cost', hue='Transportation type', data=grouped_transport)
ax.set_title('Average Transportation Cost by City and Type', fontsize=50)  
ax.set_xlabel('City', fontsize=40) 
ax.set_ylabel('Average Transportation Cost', fontsize=35)  
ax.tick_params(axis='x', labelsize=35, rotation=60)  
ax.tick_params(axis='y', labelsize=35) 
ax.legend(title='Transportation Type', title_fontsize='30', fontsize='30')  

ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.tight_layout()
plt.show()
```

```python
# Grouping data by city and accommodation type
grouped_accommodation = travel.groupby(['Destination_City', 'Accommodation type']).mean()['Accommodation cost'].reset_index()

plt.figure(figsize=(50, 20))

# Accommodation cost by city and type
ax = sns.barplot(x='Destination_City', y='Accommodation cost', hue='Accommodation type', data=grouped_accommodation)
ax.set_title('Average Accommodation Cost by City and Type', fontsize=50)  
ax.set_xlabel('City', fontsize=40) 
ax.set_ylabel('Average Accommodation Cost', fontsize=35)  
ax.tick_params(axis='x', labelsize=35, rotation=60)  
ax.tick_params(axis='y', labelsize=35) 
ax.legend(title='Accommodation Type', title_fontsize='30', fontsize='30')  

ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.tight_layout()
plt.show()
```

```python
# Grouping data by city, accommodation type, and transportation type
grouped_data = travel.groupby(['Destination_City', 'Accommodation type', 'Transportation type']).mean()['per_capita_cost'].reset_index()

# Creating a pivot table for the stacked bar chart
pivot_data = grouped_data.pivot_table(index='Destination_City', 
                                      columns=['Accommodation type', 'Transportation type'], 
                                      values='per_capita_cost', 
                                      fill_value=0)

# Plotting the stacked bar chart
plt.figure(figsize=(50, 20))
pivot_data.plot(kind='bar', stacked=True, figsize=(50, 20))

plt.title('Average Per Capita Cost by City, Accommodation Type, and Transportation Type', fontsize=50)
plt.xlabel('City', fontsize=40)
plt.ylabel('Average Per Capita Cost', fontsize=35)
plt.xticks(rotation=60, fontsize=35)
plt.yticks(fontsize=35)
plt.legend(title='Accommodation & Transportation Type', title_fontsize='20', fontsize='20')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(20, 30))  
pivot_data.plot(kind='barh', stacked=True, figsize=(20, 30)) 

plt.title('Average Per Capita Cost by City, Accommodation Type, and Transportation Type', fontsize=25)
plt.xlabel('Average Per Capita Cost', fontsize=20)
plt.ylabel('City', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(title='Accommodation & Transportation Type', title_fontsize='18', fontsize='15', loc='upper right')

plt.show()
```

# Average Costs by Time

```python
import datetime as dt

# Converting dates and creating Month-Year
travel['Start date'] = pd.to_datetime(travel['Start date'])
travel['End date'] = pd.to_datetime(travel['End date'])
travel['Month-Year'] = travel['Start date'].dt.strftime('%Y-%m')

monthly_trends = travel.groupby('Month-Year').mean()[['Accommodation cost', 'Transportation cost', 'per_capita_cost']]
monthly_trends.reset_index(inplace=True)

plt.figure(figsize=(10, 5))
plt.subplot(1, 1, 1)
sns.lineplot(x='Month-Year', y='Accommodation cost', data=monthly_trends)
plt.title('Average Accommodation Cost Trend Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Average Cost')
plt.xticks(rotation=90)  

plt.show()
```

```python
#Plot for Transportation cost trend
plt.figure(figsize=(10, 5))
plt.subplot(1, 1, 1)
sns.lineplot(x='Month-Year', y='Transportation cost', data=monthly_trends)
plt.title('Average Transportation Cost Trend Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Average Cost')
plt.xticks(rotation=90)  

plt.show()
```

```python
# Plot for Per Capita cost trend
plt.figure(figsize=(10, 5))
plt.subplot(1, 1, 1)
sns.lineplot(x='Month-Year', y='per_capita_cost', data=monthly_trends)
plt.title('Average Per Capita Cost Trend Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Average Cost')
plt.xticks(rotation=90)  

plt.show()
```

# Average Costs by Age Group

```python
bins = [0, 18, 30, 40, 50, 60, 100]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61+']
travel['age_group'] = pd.cut(travel['Traveler age'], bins=bins, labels=labels, right=False)
age_group_data = travel.groupby('age_group').mean()[['Accommodation cost', 'Transportation cost', 'per_capita_cost']]
```

```python
# Accommodation cost by age group
plt.figure(figsize=(8, 5))
plt.subplot(1, 1, 1)
sns.barplot(x=age_group_data.index, y=age_group_data['Accommodation cost'])
plt.title('Average Accommodation Cost by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Cost')

plt.show()
```

```python
# Transportation cost by age group
plt.figure(figsize=(8, 5))
plt.subplot(1, 1, 1)
sns.barplot(x=age_group_data.index, y=age_group_data['Transportation cost'])
plt.title('Average Transportation Cost by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Cost')

plt.show()
```

```python
# Per capita cost by age group
plt.figure(figsize=(8, 5))
plt.subplot(1, 1, 1)
sns.barplot(x=age_group_data.index, y=age_group_data['per_capita_cost'])
plt.title('Average Per Capita Cost by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Cost')

plt.show()
```

 # Accommodation Type and Nationality

```python
contingency_table = pd.crosstab(travel['Traveler nationality'], travel['Accommodation type'])
plt.figure(figsize=(15, 10))
sns.heatmap(contingency_table, annot=True, cmap='coolwarm', fmt='d')
plt.title('Accommodation Type by Traveler Nationality (Heatmap)')
plt.xlabel('Accommodation Type')
plt.ylabel('Traveler Nationality')

plt.show()
```

# Transportation Type and Nationality

```python
contingency_table1 = pd.crosstab(travel['Traveler nationality'], travel['Transportation type'])
plt.figure(figsize=(15, 10))
sns.heatmap(contingency_table1, annot=True, cmap='coolwarm', fmt='d')
plt.title('Transportation Type by Traveler Nationality (Heatmap)')
plt.xlabel('Transportation Type')
plt.ylabel('Traveler Nationality')

plt.show()
```
